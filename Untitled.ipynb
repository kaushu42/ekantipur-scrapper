```python
import bs4
import urllib.request as u
from bs4 import BeautifulSoup
```

```python
url = 'http://www.ekantipur.com/nep/politics'
data = u.urlopen(url).read().decode()
contents = BeautifulSoup(data, 'html.parser')
```

```python
titles = []
h1 = contents('h1')
for i in h1:
    d = i.findChildren("a" , recursive=False)
    titles.append(d[0].text)
```

```python
links = []
divs = contents('div', {'class': 'col-xs-12 col-sm-12 col-md-12 text-wrap'})
for div in divs:
    a = div.findChildren('a')
    links.append(a[0]['href'])
```

```python
with open('titles.txt', 'w') as f:
    for title in titles:
        f.write(title + '\n')
```

```python
with open('links.txt', 'w') as f:
    for link in links:
        f.write(link + '\n')
```

```python
with open('links.txt', 'r') as f:
    print(f.read())
```

```python
links[0], links[1]
```

```python
file = open('news.txt', 'w')
for i in range(len(titles)):
    url = links[i]
    news_data = u.urlopen(url).read().decode()
    news_content = BeautifulSoup(news_data, 'html.parser')
    
    div = news_content('div')
    
    for x in div:
        try:
            if x.attrs['id'] == 'content':
                ourTag = x
                break
        except:
            pass
    
    pTags = ourTag.findAll('p')
    for k in pTags:
        for j in k.children:
            if type(j) == bs4.element.NavigableString:
                file.write(j)
    print(f'Done {i}')
    file.write('\n*********************************\n')
file.close()
```
