# Import the libraries

```python
import bs4
import urllib.request as u
from bs4 import BeautifulSoup
```

# Download the main web page
Looks at the main page of ekantipur topics(sports, world, Nepal, etc.)

```python
topic = 'world'
url = 'http://www.ekantipur.com/nep/' + topic
data = u.urlopen(url).read().decode()
contents = BeautifulSoup(data, 'html.parser')
```

# This collects the titles and links of each story
We will go into each link and download the content later on

```python
# Get the titles
titles = []
h1 = contents('h1') # Titles are inside h1/a/
for i in h1:
    d = i.findChildren("a" , recursive=False)
    titles.append(d[0].text)
```

```python
# Get the links of each story
links = []
divs = contents('div', {'class': 'col-xs-12 col-sm-12 col-md-12 text-wrap'}) # Links are inside this type of div
for div in divs:
    a = div.findChildren('a')
    links.append(a[0]['href'])
```

# Write the scraped titles and links to files

```python
with open(topic + '/titles.txt', 'w') as f:
    for title in titles:
        f.write(title + '\n')
```

```python
with open(topic + '/links.txt', 'w') as f:
    for link in links:
        f.write(link + '\n')
```

## View the saved data

```python
# Viewing only a single line due to f.readline(). If you want to see the full file, replace it with f.read()
with open(topic + '/titles.txt', 'r') as f:
    print(f.readline())
    
with open(topic + '/links.txt', 'r') as f:
    print(f.readline())
```

# Go inside each link and collect the content of each link

```python
# Open a file to write news data
file = open(topic + '/news.txt', 'w')

print('Done: ', end='')

# Repeat for all links
for i in range(len(titles)):
    url = links[i]
    news_data = u.urlopen(url).read().decode()
    news_content = BeautifulSoup(news_data, 'html.parser')
    
    div = news_content.find(id='content')
   
    # The <p> tags contain all the information
    pTags = div.findAll('p')
    
    for k in pTags:# Look inside each <p> tag
        for j in k.children: # Check the children of each <p> tag
            if type(j) == bs4.element.NavigableString:  # Save the data if the child is a String. Else, it is information
                                                        # we don't need.
                file.write(j)
    print(i + 1, end=', ')
    file.write('\n*********************************\n')
file.close()
```
