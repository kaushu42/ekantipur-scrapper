```python
import pandas as pd
import pickle
```

```python
topics = [
    'economy',
    'entertainment',
    'nepal',
    'others',
    'politics',
    'sports',
    'world',
]
```

```python
def read_data(filename):
    return open(filename, 'r').read()
```

```python
titles_all = []
news_all = []

for i in topics:
    print(f'In topic: {i}')
    
    titles = read_data(i + '/titles.txt')
    titles = titles.split('\n')[:-1]
    print(f'Total Titles: {len(titles)}')
    
    news = read_data(i + '/news.txt')
    news = news.split('\n*********************************\n')[:-1] # Last item is empty
    print(f'Total News: {len(news)}')
    
    titles_all += titles
    news_all += news
    print()
```

```python
df = pd.DataFrame(data={'title':titles_all, 'data':news_all})
df
```

# Pickle the data to disk

```python
pickle.dump(df, open('data.pickle', 'wb'))
```

# Read the pickled data

```python
pickle.load(open('data.pickle', 'rb'))
```
